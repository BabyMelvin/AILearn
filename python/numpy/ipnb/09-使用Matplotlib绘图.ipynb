###  9.2绘制多项式函数

`polyld`来创建多项式。

```python
# 1.自然数序列为多项式系数，使用polyld函数创建多项式
func=poly1d(array([1,2,3,4]).astype(float))
print(func)
```

```python
# 2.使用Numpy的linespace函数创建x轴数值，-10到10之间产生30个均匀分布值
x=linspace(-10,10,30)
print(x)
```

```python
# 3.计算我们第一步中创建的多项式的值
y=func(x)
# 4.调用plot函数，这并不会立即显示函数
plot(x,y)
# 5.使用xlabel函数添加x轴标签
xlabel('x')
# 6.使用ylabel函数添加y轴标签
ylabel('y(x)')
# 7.使用show函数显示图像
show()
```

###  9.3格式字符串

plot函数可以接受任意个参数。还可以选用指定颜色和风格。

* x,y:数组或标量。数据点的水平或竖直坐标。x，是可选的，没有默认为`[0,...,N-1]`
* fmt:字符串，可选的。`ro`红色圆圈。`fmt='[color][marker][line]'`

**颜色**

|字符|color|
|--|--|
|'b'|blue|
|'g'|green|
|'r'|red|
|'c'|cyan(l蓝绿色)|
|'m'|magenta洋红|
|'y'|yellow|
|'k’|black|
|'w'|white|

**标记**

|字符|描述|
|--|--|
|'.'|点标记|
|','|像素标记|
|'o'|圆圈标记|
|'^'|正三角|
|'V'|反三角|
|'<'|左三角|
|'>'|右三角|
|'s’|波浪|
`*`,`+`,`x`等

**线类型**

|符号|描述|
|--|--|
|'-'|实线|
|'--'|虚线|
|'-.'|点划线|
|':'|点线|

* lines:2维对象数据表示画的点。
* `**kwargs`二维属性。




```python
#　1.创佳多项式及导函数(ploy 1(数字) d)
func=poly1d(array([1,2,3,4]).astype(float))
func1=func.deriv(m=1)
x=linspace(-10,10,30)
y=func(x)
y1=func1(x)
# 不同风格绘制：红色圆形和绿色虚线
plot(x,y,'ro',x,y1,'g--')
xlabel('x')
ylabel('y')
show()
```

###  9.5子图

利用subplot函数完成，来分组完成。

```python
# 1.创建多项式函数及其导函数
func=poly1d(array([1,2,3,4]).astype(float))
x=linspace(-10,10,30)
y=func(x)
func1=func.deriv(m=1)
func2=func.deriv(m=2)
y1=func1(x)
y2=func2(x)

# 2.画第一个子图。标题设置为Polynomial，红实线绘制
subplot(311)
plot(x,y,'r-')
title('polynomial')

# 3.第二个子图。蓝色三角形
subplot(312)
plot(x,y1,'b^')
title('first derivative')

# 4.第三个子图，绿色圆形
subplot(313)
plot(x,y2,'go')
title('second derivative')
xlabel('x')
ylabel('y')
show()
```

## 9.9直方图
直方图(histogram)可将数据分布可视化。`hist`函数可以绘制直方图。
参数有两项-数据的数组和柱形的数量。

```python
x=array([12,2,1,5,3,2,3,15,20,3,4,1,3,13])
hist(x,bins=[0,2,4,8,10,12,14,16,18,20])
title('histogram')
show()
```

### 9.13散点图

scatter plot，可指定颜色和大小，以及图像的透明度

```python
# basic
figure(1)
x=rand(50,30)
subplot(211)
scatter(x[:,1],x[:,2])

# with label 
subplot(212)
# label设定为[1:20]->1, [21:35]->2, [36:50]->3，
#（python中数组连接方法：先强制转为list，用+，再转回array）
label=list(ones(20))+list(2*ones(15))+list(3*ones(15))
label=array(label)
print(15.0*label)
scatter(x[:,1],x[:,0],15.0*label,15*label)

# with legend
figure(2)
idx_1=find(label==1)
p1=scatter(x[idx_1,1],x[idx_1,0],marker='x',color='m',label='1',s=30)
idx_2=find(label==2)
p2=scatter(x[idx_2,1],x[idx_2,0],marker='+',color='c',label='2',s=100)
idx_3=find(label==3)
scatter(x[idx_3,1],x[idx_3,0],marker='o',color='r',label='3',s=50)
legend(loc='upper right')
show()
```
