```python
t1=arange(0,5,0.1)
t2=arange(0,5,0.02)

def f(t):
    # exp e为低的指数函数
    return exp(-t)*cos(2*pi*t)
figure(1)
subplot(211)
plot(t1,f(t1),'bo',t2,f(t2),'k')

subplot(212)
plot(t2,cos(2*pi*t2),'r--')
show()
```

```python
# fix random state for reproducibility
# random.seed(19680801)

# make up some data in the interal [0,1]
y=normal(loc=0.5,scale=0.4,size=1000)
y=y[(y>0)&(y<1)]
y.sort()
x=arange(len(y))

# plot with various axes scales
figure(1)

# linear
subplot(221)
plot(x,y)
ylabel('linear')
title('linear')
grid(True)

# log
plt.subplot(222)
plot(x,y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
subplot(223)
plot(x,y-y.mean())
yscale('symlog',linthreshy=0.01)
title('symlog')

# logit
subplot(224)
plot(x,y)
yscale('logit')
title('logit')
grid(True)

# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)
```

```python
# 多个figs
t=arange(0,2,0.01)
s1=sin(2*pi*t)
s2=sin(4*pi*t)

figure(1)
subplot(211)
plot(t,s2,'s')

# Get the current Axes instance 
ax=gca()
ax.set_xticklabels([])
subplot(212)
plot(t,2*s1)

figure(2)
plot(t,s2)
show()
```

```python
t1=arange(0,5,0.1)
t2=arange(0,5,0.02)
t3=arange(0,2,0.01)

def f(t):
    s1=cos(2*pi*t)
    e1=exp(-t)
    return s1*e1
subplot(121)
plot(t1,f(t1),'o',t2,f(t2),'-')
title('subplot 1')
ylabel('damped oscilation')
suptitle('this is a somewhat long figure title',fontsize=20)

subplot(122)
plot(t3,cos(2*pi*t3),'--')
xlabel('time(s)')
title('subplot 2')
ylabel('undamped')

# 调整图形位置
subplots_adjust(left=0.2,wspace=0.8,top=0.8)
show()
```
