## pylab模块
包含许多Numpy和pyplot常用函数模块，方便快速计算和绘图。`import pylab as pl`

```python
import matplotlib.pyplot as plt
x =  np.linspace(0,10,1000)
y=np.sin(x)
z=np.cos(x**2)
# 当前Figure对象
plt.figure(figsize=(8,4))
# 标签$$内嵌LaTex引擎
plt.plot(x,y,label="$sin(x)$",color="red",linewidth=2)
plt.plot(x,z,"b--",label="$cos(x^2)$")
plt.xlabel("Time(s)")
plt.ylabel("Volt")
# 显示范围
plt.ylim(-1,2,1.2)
# 显示图例
plt.legend()
plt.show()
#　保存图片，关闭无法保存
plt.savefig("test.png",dpi=120)
```

```python
# 保存图片并获得其中的字节数据
import io
buf=io.BytesIO()
# 将图像以png格式保存子啊buf中
plt.savefig(buf,fmt="png")
buf.getvalue()[:20]
```
